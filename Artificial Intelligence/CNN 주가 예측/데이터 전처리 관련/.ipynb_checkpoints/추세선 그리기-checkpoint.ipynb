```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import linregress
import FinanceDataReader as fdr
import mplfinance as mpf
code = "004000"
start = "2019-12-01"
end = "2020-01-01"
df = fdr.DataReader(code, start, end)
df = df[["High", "Low"]]

x = df["High"]
y = df["Low"]

# plot the data itself
plt.plot(x,y,'o')

# calc the trendline
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x,p(x),"r--")
```

```python
code = "004000"
start = "2019-12-01"
end = "2020-01-01"
df = fdr.DataReader(code, start, end)
df = df[["Open", "High", "Low", "Close", "Volume"]]

rc={
    "axes.labelcolor": "none",
    "axes.spines.bottom": False,
    "axes.spines.left": False,
    "axes.spines.right": False,
    "axes.spines.top": False,
    "font.size": 0,
    "xtick.color": "none",
    "ytick.color": "none",
}
mc=mpf.make_marketcolors(up='red', down='blue', alpha=1.0, inherit=True)
codes = ["004000", "063170", "009520", "121800", "064260"]
style_final=mpf.make_mpf_style(marketcolors=mc, rc=rc, facecolor = "white", figcolor='white')

mpf.plot(df, type='candle', style=style_final)
```
