```python
import tensorflow as tf

from tensorflow.keras.layers import Dense, Conv1D, LSTM, Input, TimeDistributed
from tensorflow.keras.models import Model
```

```python
# 32는 batch, 25는 time_step, 1은 feature 개수.
x = tf.random.uniform(shape=(32, 25, 1))
x
```

# return_sequences=False 인 경우

```python
lstm = LSTM(20)
output_a = lstm(x)

# shape에 32는 batch 개수, 20은 위에서 lstm에 지정한 output 개수.
output_a.shape
```

# return_sequences=True 인 경우

```python
lstm = LSTM(20, return_sequences=True)
output_b = lstm(x)
output_b.shape
```

# return_sequence=False 를 Dense에 넘겨줄 경우

```python
dense = Dense(10)
dense(output_a).shape
```

# return_sequence=True 를 Dense에 넘겨줄 경우

```python
dense = Dense(10)
dense(output_b).shape
```

# TimeDistributed layer 활용

```python
time = TimeDistributed(Dense(10))
time(output_b).shape
```

# cell state와 hidden layer

```python
lstm = LSTM(20, return_sequences=True, return_state=True)
output_c, _hidden, _state = lstm(x)
```
