```python
import numpy as np
import math

x = np.linspace(2, math.pi, 2)
y = np.sin(x)
print(x)
print(y)
a = np.random.randn()
b = np.random.randn()
c = np.random.randn()
d = np.random.randn()
print("{} {} {} {}".format(a,b,c,d))
lr = 0.0001

# x에 대한 y 값 예측.
for t in range(1000):
    y_pred = a + b * x + c * x ** 2 + d * x ** 3 # y = a + bx + cx^2 + dx^3 (예측값 계산)
    loss = np.square(y_pred - y).sum() # 오차 계산

    # 가중치에대해서 오차를 곱한값으로 갱신시키는데, 발산을 방지하기 위해서, learning rate를 곱해서 갱신해 준다.
    # 이걸 반복 할 수록, loss값은 낮아지고, target값에 근접할 수 있는, 가중치를 구할 수 있음.
    grad_y_pred = 2.0 * (y_pred - y)
    grad_a = grad_y_pred.sum()
    grad_b = (grad_y_pred * x).sum()
    grad_c = (grad_y_pred * x ** 2).sum()
    grad_d = (grad_y_pred * x ** 3).sum()
    if t & 100 == 0:
        print("{} | {:.5f} {:.5f} {:.5f} {:.5f}  // y_pred : {}    loss : {}".format(t,a,b,c,d,y_pred,loss))
        print("기울기 : ",grad_y_pred, "\n")
    a -= lr * grad_a
    b -= lr * grad_b
    c -= lr * grad_c
    d -= lr * grad_d

print("{} {} {} {}".format(a,b,c,d))
```
